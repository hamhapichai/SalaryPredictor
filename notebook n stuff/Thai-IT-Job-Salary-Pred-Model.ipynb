# Data Preprocessing

## Import Data form CSV

```python
df = pd.read_csv('thai-it-salary-form-response-2022.csv')

df = df[['WorkPlace', 'EdLevel', 'YearsCodePro', 'YearsCode', 'Employment',
         'WorkPosition', 'DevType', 'OrgSize', 'Salary']]
```

```python
print(df.shape)
df
```

## Education

```python
df['EdLevel'].value_counts()
```

```python
def clean_education(x):
  if 'Bachelor' in x or 'Certificate' in x:
    return 1
  if 'Master' in x:
    return 2
  if 'Doctoral' in x:
    return 2
  return 0

df['EdLevel'] = df['EdLevel'].apply(clean_education)
df['EdLevel'].value_counts()
```

## Employment

```python
df['Employment'].value_counts()
```

```python
def clean_employment(x):
  if 'full-time' in x or 'freelance' in x:
    return 1
  if 'part-time' in x:
    return 0
  return 1

df['Employment'] = df['Employment'].apply(clean_employment)
df['Employment'].value_counts()
```

## OrgSize

```python
df['OrgSize'].value_counts()
```

```python
def clean_orgsize(x):
  if 'Just me' in x or '10 to 19' in x or '2 to 9' in x:
    return 0
  if '20 to 99' in x:
    return 1
  if '100 to 999' in x:
    return 2
  if '1,000 to 9,999' in x:
    return 3
  if '10,000 or more':
    return 4
  return x

df['OrgSize'] = df['OrgSize'].apply(clean_orgsize)
df['OrgSize'].value_counts()
```

## States

```python
def get_in(x):
  jobs = []
  substring = ''
  iscollecting = False
  for s in x:
    if iscollecting:
      if s == ')':
        jobs.append(substring)
        substring = ''
        iscollecting = False
      else: substring += s

    if s == '(':
      iscollecting = True
  return jobs[0]
```

```python
df['WorkPlace'].unique()
```

```python
def clean_state(x):
  xstr = get_in(x)
  if xstr == 'Bangkok' or xstr == 'Nonthaburi' or xstr == 'Samut Prakan':
    return 1
  else:
    return 0

df['WorkPlace'] = df['WorkPlace'].apply(clean_state)
```
